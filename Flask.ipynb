```python
templates = {
    "templates/base.html": """
<!DOCTYPE html> 
<html lang="en"> 
<head> 
    <meta charset="UTF-8"> 
    <title>{% block title %}Activity 10{% endblock %}</title> 
    <link rel="stylesheet" href="{{ url_for('static', filename='css/style.css')}}"> 
</head> 
<body> 
    <header> 
        <img src="{{ url_for('static', filename='images/TUP.png') }}" alt="TUP Logo"> 
        <h1>Activity 10 Flask Web App</h1>
    </header> 
    <nav> 
        <a href="{{ url_for('home') }}">Home</a> 
        <a href="{{ url_for('about') }}">About</a> 
        <a href="{{ url_for('contact') }}">Contact</a> 
        {% if current_user.is_authenticated %} 
            <a href="{{ url_for('students') }}">Students</a> 
            <a href="{{ url_for('logout') }}">Logout</a> 
        {% else %} 
            <a href="{{ url_for('login') }}">Login</a> 
            <a href="{{ url_for('register') }}">Register</a> 
        {% endif %} 
    </nav> 
    <main> 
        {% with messages = get_flashed_messages() %} 
            {% if messages %} 
                {% for msg in messages %} 
                    <p style="color: red;">{{ msg }}</p> 
                {% endfor %} 
            {% endif %} 
        {% endwith %} 
        {% block content %}{% endblock %} 
    </main> 
</body> 
</html>""",

    "templates/home.html": """
{% extends 'base.html' %} 
{% block content %} 
<h2>Welcome to the Home Page!</h2> 
<p>This is a simple Flask web application.</p> 
{% endblock %}""",

    "templates/about.html": """
{% extends 'base.html' %} 
{% block title %}About{% endblock %} 
{% block content %} 
<h2>About This App</h2> 
<p>This Flask app demonstrates user login and database integration.</p> 
{% endblock %}""",

    "templates/contact.html": """
{% extends 'base.html' %} 
{% block content %} 
<h2>Contact Us</h2> 
<p>Email: gilfredallen_madrigal@tup.edu.ph</p> 
{% endblock %}""",

    "templates/login.html": """
{% extends 'base.html' %} 
{% block content %} 
<h2>Login</h2> 
<form method="POST"> 
    {{ form.hidden_tag() }} 
    <p>{{ form.email.label }}<br>{{ form.email() }}</p> 
    <p>{{ form.password.label }}<br>{{ form.password() }}</p> 
    {{ form.submit() }} 
</form> 
{% endblock %}""",

    "templates/register.html": """
{% extends 'base.html' %} 
{% block content %} 
<h2>Register</h2> 
<form method="POST"> 
    {{ form.hidden_tag() }} 
    <p>{{ form.email.label }}<br>{{ form.email() }}</p> 
    <p>{{ form.password.label }}<br>{{ form.password() }}</p> 
    {{ form.submit() }} 
</form> 
{% endblock %}""",

    "templates/students.html": """
{% extends 'base.html' %} 
{% block title %}Students{% endblock %} 
{% block content %} 
<h2>Student List</h2> 
<form method="POST" action="{{ url_for('add_student') }}"> 
    <input type="text" name="name" placeholder="Full Name" required> 
    <input type="email" name="email" placeholder="Email" required> 
    <button type="submit">Add Student</button> 
</form> 
<ul> 
    {% for student in students %} 
        <li>{{ student.full_name }} ({{ student.email }}) - 
            <a href="{{ url_for('delete_student', id=student.id) }}">Delete</a> 
        </li>
    {% endfor %}
</ul>
{% endblock %}"""
}

# Write all files
for path, content in templates.items():
    with open(path, "w") as f:
        f.write(content.strip())

print("All HTML templates saved successfully!")
```

```python
# --- PART 1: Update the Template File ---
home_html_custom = """
{% extends 'base.html' %}
{% block content %}
<h2>Hello, I am {{ Mark Aldred }}!</h2>
<p>Welcome to my Flask Activity 10 web page.</p>
<p>Section: {{ BSECE-1b }}</p>
{% endblock %}
"""

# Ensure the directory exists first
os.makedirs("templates", exist_ok=True)

with open("templates/home.html", "w") as f:
    f.write(home_html_custom.strip())

# --- PART 2: The Flask Application Code ---
from flask import Flask, render_template
app = Flask(__name__)

@app.route('/')
def home():
    # This sends 'name' and 'section' to the HTML file we just created above
    return render_template('home.html', name="Mark Aldred", section="BSECE-1b")
    
# This part is just for Jupyter/Colab to help it run correctly
if __name__ == '__main__':
    print("Template updated and server starting...")
    # app.run()  # Uncomment this when you are ready to launch!
```

```python
from app import app 
from models import db, Student 
with app.app_context(): 
    # Add a new student 
    sample_student = Student(full_name="Maria Santos", 
email="maria@example.com") 
    db.session.add(sample_student) 
    db.session.commit() 
    # Fetch and display students 
    all_students = Student.query.all() 
    for s in all_students: 
        print(f"{s.id}: {s.full_name} ({s.email})")
```
